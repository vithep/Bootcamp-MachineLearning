# 3. Preparando o Dataset (Categorico)

#### Importação das Bibliotecas

```python
import pandas as pd
import numpy as np
import datetime
```

#### Realizando a leitura do arquivo (dataset)

```python
df = pd.read_csv('imports-85.txt', sep=',',header=None,na_values="?")
```

#### Apresentando as 5 primeiras linhas do dataset

```python
df.head()
```

```python
header = {0:"symboling", 1:"normalized-losses", 2:"make", 3:"fuel-type", 4:"aspiration",
          5:"num-of-doors", 6:"body-style", 7:"drive-wheels", 8:"engine-location",
          9:"wheel-base", 10:"length", 11:"width", 12:"height", 13:"curb-weight",
         14:"engine-type", 15:"num-of-cylinders", 16:"engine-size", 17:"fuel-system",
         18:"bore", 19:"stroke", 20:"compression-ratio", 21:"horsepower", 22:"peak-rpm",
         23:"city-mpg", 24:"highway-mpg", 25:"price"}
```

```python
df.rename(columns=header,inplace=True)
df.head()
```

#### Verificando os tipos de variáveis e se existem ou não valores nulos

```python
df.info()
```

#### Conta a quantidade de object

```python
df.dtypes.eq('object').sum()
```

#### Selecionando apenas os dados categoricos

```python
df2 = df.select_dtypes(include=['object']).copy()  #realiza uma cópia do dataset apenas para as colunas do tipo string
df2.head()
```

```python
df2.info()
```

#### Contando valores nulos

```python
df2.isnull().sum()
```

#### Realiza o drop para valores nulos - no caso aqui, das linhas

```python
df2.dropna(inplace=True)
df2.info()
```

#### Conta os valores para cada uma das classes

```python
df2['num-of-cylinders'].value_counts()
```

#### Transformando em numeros ( replace )

```python
mapeamento_classes = {"num-of-doors":{"four":4, "two":2},
                   "num-of-cylinders":{"four":2, "six":6, "five":5, "eight":8, "two":2, "twelve": 12, "three":3}}
```

#### Aplicando o replace

```python
df2.replace(mapeamento_classes,inplace=True)
df2.head()
```

#### Aplicando o label encoding 

```python
df2['body-style'] = df2['body-style'].astype('category')
df2.dtypes
```

#### Procedimento label encoding 

```python
df2['body-style-cat']=df2['body-style'].cat.codes
df2.head()
```

```python
df2.info()
```

```python
df2['body-style-cat'].value_counts()
```

#### Aplicando o one-hot encoding

```python
df2['drive-wheels'].unique()
```

#### Aplicando o hot enconding com o panda

```python
pd.get_dummies(df2, columns=['drive-wheels']).head()
```
